# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def fun(n):
    count = 0
    while n != 0:
        nam = n % 2
        if nam ==  1:
            count +=1
        n = n // 2
    return count

def run_tests():
    test_cases = [
        (0, 0, "Число 0"),
        (1, 1, "Число 1"),
        (2, 1, "Число 2"),
        (3, 2, "Число 3"),
        (5, 2, "Число 5"),
        (7, 3, "Число 7"),
        (255, 8, "Число 255"),
    ]
    
    passed = 0
    for num, expected, desc in test_cases:
        result = fun(num)
        if result == expected:
            print(f"✓ {desc}: {result} == {expected}")
            passed += 1
        else:
            print(f"✗ {desc}: {result} != {expected}")
    
    print(f"\nРезультат: {passed}/{len(test_cases)} тестов пройдено")

if __name__ == "__main__":
    run_tests()
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n):
    count = 0
    while n >= 10:
        product = 1
        for digit in str(n):
            product *= int(digit)
        n = product
        count += 1
    return count

def run_tests():
    test_cases = [
        (4, 0),   # уже однозначное
        (9, 0),   # уже однозначное
        (39, 3),  # 39 -> 27 -> 14 -> 4
        (999, 4), # 999 -> 729 -> 126 -> 12 -> 2
        (10, 1),  # 10 -> 0
        (25, 2),  # 25 -> 10 -> 0
        (77, 4),  # 77 -> 49 -> 36 -> 18 -> 8
        (123, 1), # 123 -> 6
        (0, 0),   # ноль
    ]
    
    passed = 0
    for num, expected in test_cases:
        result = magic(num)
        if result == expected:
            print(f"✓ magic({num}) = {result}")
            passed += 1
        else:
            print(f"✗ magic({num}) = {result}, ожидалось {expected}")
    
    print(f"\nРезультат: {passed}/{len(test_cases)} тестов пройдено")

if __name__ == "__main__":
    run_tests()
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(pred, true):
    
    if len(pred) == 0:
        return 0.0
    
    total = 0
   
    for i in range(len(pred)):
        total += (pred[i] - true[i]) ** 2
    
    return total / len(pred)



def run_tests():
    tests = [
        ([1, 2, 3], [1, 2, 3], 0),
        ([1, 2, 3], [2, 3, 4], 1),
        ([0, 0, 0], [1, 1, 1], 1),
        ([1, 2], [3, 4], 4),
        ([5], [10], 25),
    ]
    
    print("Простые тесты")
    for pred, true, expected in tests:
        result = mse(pred, true)
        if result == expected:
            print(f"✓ mse({pred}, {true}) = {result}")
        else:
            print(f"✗ mse({pred}, {true}) = {result}, ожидалось {expected}")

if __name__ == "__main__":
    run_tests()
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def prime_factors(n):
    
    result = ""
    i = 2
    while i * i <= n:
        count = 0
        while n % i == 0:
            count += 1
            n //= i
        if count == 1:
            result += f"({i})"
        elif count > 1:
            result += f"({i}**{count})"
        i += 1
    if n > 1:
        result += f"({n})"
    return result



def test():
    
    tests = [
        (86240, "(2**5)(5)(7**2)(11)"),
        (2, "(2)"),
        (4, "(2**2)"),
        (6, "(2)(3)"),
        (12, "(2**2)(3)"),
        (1, ""),
    ]
    
    for num, expected in tests:
        result = prime_factors(num)
        if result == expected:
            print(f"✓ {num} -> {result}")
        else:
            print(f"✗ {num} -> {result} (ожидалось {expected})")


if __name__ == "__main__":
    test()
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(n):
    
    total = 0
    k = 0
    while total < n:
        k += 1
        total += k * k
        if total == n:
            return k
    return "It is impossible"



def test():
    tests = [
        (1, 1),      # Можно: 1 = 1^2
        (5, 2),      # Можно: 5 = 1^2 + 2^2
        (14, 3),     # Можно: 14 = 1^2 + 2^2 + 3^2
        (30, 4),     # Можно: 30 = 1^2 + 2^2 + 3^2 + 4^2
        (2, "It is impossible"),  # Нельзя
        (3, "It is impossible"),  # Нельзя
        (4, "It is impossible"),  # Нельзя
    ]
    
    for n, expected in tests:
        result = pyramid(n)
        if result == expected:
            print(f"✓ pyramid({n}) = {result}")
        else:
            print(f"✗ pyramid({n}) = {result}, ожидалось {expected}")

if __name__ == "__main__":
    test()
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced(n):
    
    s = str(n)
    length = len(s)
    
    if length % 2 == 0:
        mid = length // 2
        left_sum = sum(int(d) for d in s[:mid-1])
        right_sum = sum(int(d) for d in s[mid+1:])
    else:
        mid = length // 2
        left_sum = sum(int(d) for d in s[:mid])
        right_sum = sum(int(d) for d in s[mid+1:])
    
    return left_sum == right_sum



def test():
    tests = [
        (1, True),      # Однозначное
        (11, True),     # 1 == 1
        (12, False),    # 1 != 2
        (131, True),    # 1 == 1
        (123, False),   # 1 != 3
        (12321, True),  # 1+2 == 2+1
        (1001, True),   # 1 == 1
    ]
    
    for n, expected in tests:
        result = is_balanced(n)
        if result == expected:
            print(f"✓ is_balanced({n}) = {result}")
        else:
            print(f"✗ is_balanced({n}) = {result}, ожидалось {expected}")

if __name__ == "__main__":
    test()
```
